```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import wesanderson
from matplotlib.legend_handler import HandlerTuple
```

```python
def load_df(file):
    df = pd.read_csv(file)
    
    id_regex = r'mouse\s*(\d+)'
    gender_regex = r'([FM])'
    treatment = df['Image'].apply(lambda x: 'rLP' in x)
    
    df['ID'] = df['Image'].str.extract(id_regex)
    df['Gender'] = df['Image'].str.extract(gender_regex)
    df['Treatment'] = np.where(treatment, 'rLP', 'control')

    return df  
```

```python
def process_replicates(df):
    """
    Identifies technical replicates and processes them. Average numerical values, and copy over metadata, which should be matching between the samples.
    
    Technical replicates: datapoints with matching sample #s, repeated measurements at same timepoint. Average values.
    """
    
    grouped = df.groupby(['ID', 'Gender', 'Treatment'])

    # identify technical replicates
    technical_replicates = [group for _, group in grouped if len(group) > 1]
    
    for replicate_pair in technical_replicates:

        # average area, D54D2, and Iba1
        avg_rep = replicate_pair.mean(numeric_only=True)

        # copy over metadata
        rep1 = replicate_pair.iloc[0]
        avg_rep['Image'], avg_rep['ID'], avg_rep['Gender'], avg_rep['Treatment'] = rep1['Image'], rep1['ID'], rep1['Gender'], rep1['Treatment']
        avg_rep = pd.DataFrame([avg_rep])

        # update rows in df
        df = df[df['ID'] != rep1['ID']]
        df = pd.concat([df, avg_rep], ignore_index=True)

    return df
```

```python
def boxplot(df, x, y, color, color_order, cat_order, y_range, title, x_label=None, y_label=None):
    """
    Prints a boxplot and individual data points (w/ tech reps averaged) for given df, x, y.
    """
    ax = sns.boxplot(data=df, x=x, y=y, hue=color, hue_order=color_order, order=cat_order, boxprops={'alpha': 0.4})
    sns.stripplot(data=df, x=x, y=y, hue=color, hue_order=color_order, order=cat_order, dodge=True, ax=ax)
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles=[(handles[0], handles[2]), (handles[1], handles[3])],
              labels=color_order,
              loc='upper left', handlelength=4,
              handler_map={tuple: HandlerTuple(ndivide=None)})
    plt.ylim(y_range)
    if x_label:
        ax.set_xlabel(x_label)
    if y_label:
        ax.set_ylabel(y_label)
    ax.set_title(title)
```

```python
boxplot(df, x='Treatment', y='D54D2', color='Gender', color_order=['M', 'F'], cat_order=['rLP', 'control'], 
        y_range=(1e9, 8e9), title='D54D2 by Treatment Group')
```

```python
# identify representative images for each category

def median(arr):
    """
    For lists of odd length, returns the median value.
    For lists of even length, returns the smaller of the two "middle" values.
    """
    n = len(arr)
    arr = arr.sort_values().reset_index(drop=True)
    return arr[n//2]


def identify_rep_images(df, col):
    """
    Identifies the median ld count for each subtype, e.g., apoe2 m 3mo, apoe2 f 3mo; then select from list
    """
    samples = {}

    grouped = df.groupby(['Treatment', 'Gender'])
    for _, group in grouped:
        # group by subtype
        treatment, gender = group['Treatment'].iloc[0], group['Gender'].iloc[0]
        subtype = treatment + gender

        # choose image with median LD Count
        med = median(group[col])
        idx = group.loc[group[col] == med, 'Image'].iloc[0]
        samples[subtype] = idx

    print(col + ' rep images: ')
    for k, v in samples.items():
        print(k, v)
    print('\n')


identify_rep_images(df, 'D54D2')
identify_rep_images(df, 'Iba1')
```
